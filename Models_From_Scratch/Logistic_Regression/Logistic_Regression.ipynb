```python
import numpy as np

# Feature matrix (X)  ##3 features ##3 weights will be there 
X = np.array([
    [148, 33.6, 50],
    [85, 26.6, 31],
    [183, 23.3, 32],
    [89, 28.1, 21],
    [137, 43.1, 33],
    [116, 25.6, 30],
    [78, 31.0, 26],
    [115, 35.3, 29]
])

# Target labels (y)
y = np.array([1, 0, 1, 0, 1, 0, 0, 1])
```

```python
#no of features
n_features=X.shape[1]

#weights and bias
w=np.zeros(n_features)
b=0
```

```python

#sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
z=np.dot(X,w)+b
y_pred=sigmoid(z)
```

```python
#loss function
def compute_loss(y,y_pred):
    m=y.shape[0]
    loss= -(1/m) * np.sum(y* np.log(y_pred) + (1-y)* np.log(1-y_pred))
    return loss

#predict the current loss
loss = compute_loss(y, y_pred)
print("Loss:", loss)
```

```python
##gradient descent
learning_rate = 0.0001
epochs = 1000

for i in range(epochs):
    # Forward propagation
    z = np.dot(X, w) + b
    y_pred = sigmoid(z)

    # Compute loss
    loss = compute_loss(y, y_pred)

    # Update parameters
    dw = (1 / X.shape[0]) * np.dot(X.T, (y_pred - y))
    db = (1 / X.shape[0]) * np.sum(y_pred - y)

    w = w - learning_rate * dw
    b = b - learning_rate * db

    # Print loss occasionally
    if i % 100 == 0:
        print(f"Epoch {i}, Loss: {loss}")
```

```python
#prediction
def predict(X, w, b):
    z = np.dot(X, w) + b
    y_pred= sigmoid(z)
    return (y_pred >= 0.5).astype(int)
```

```python
y_pred = predict(X, w, b)
print("Predicted labels:", y_pred)
print("Actual labels:   ", y)
```

```python
#accuracy
def accuracy(y, y_pred):
    return np.mean(y == y_pred)

acc = accuracy(y, y_pred)
print("Accuracy:", acc)
```
