```python
import math
from collections import Counter

class KNN:
    def __init__(self, K):
        self.K = K

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def _predict_single(self, test_point):
        distances = []

        for i in range(len(self.X_train)):
            distance = 0
            for j in range(len(test_point)):
                distance += (test_point[j] - self.X_train[i][j]) ** 2
            distance = math.sqrt(distance)
            distances.append((distance, self.y_train[i]))

        distances.sort(key=lambda x: x[0])
        k_neighbors = distances[:self.K]

        labels = [label for _, label in k_neighbors]
        prediction = Counter(labels).most_common(1)[0][0]

        return prediction

    def predict(self, X_test):
        predictions = []
        for point in X_test:
            predictions.append(self._predict_single(point))
        return predictions
```

```python
X_train = [[1,2],[2,3],[3,3],[6,5]]
y_train = ['Red','Red','Blue','Blue']
X_test = [[3,2],[5,4]]

knn = KNN(K=3)
knn.fit(X_train, y_train)
predictions = knn.predict(X_test)

print(predictions)
# Output: ['Red', 'Blue']
```
